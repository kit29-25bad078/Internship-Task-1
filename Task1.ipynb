```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product_A": [12000, 15000, 18000, 17000, 20000, 22000],
    "Product_B": [10000, 12000, 14000, 13000, 15000, 17000],
    "Product_C": [8000, 10000, 12000, 11000, 14000, 16000]
}

df = pd.DataFrame(data)


df["Total_Sales"] = (
    df["Product_A"] +
    df["Product_B"] +
    df["Product_C"]
)

print("========== SALES DASHBOARD ==========\n")
print(df)

print("\n========== KPI REPORT ==========")
print("Total Revenue :", df["Total_Sales"].sum())
print("Average Monthly Sales :", round(df["Total_Sales"].mean(), 2))
print("Highest Sales :", df["Total_Sales"].max())
print("Lowest Sales :", df["Total_Sales"].min())

best_month = df.loc[df["Total_Sales"].idxmax(), "Month"]
print("Best Performing Month :", best_month)

print("\n========== SUMMARY ==========")
print(df.describe())

plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Total_Sales"], marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()


plt.figure(figsize=(8,5))
plt.bar(df["Month"], df["Product_A"], label="Product A")
plt.bar(df["Month"], df["Product_B"], bottom=df["Product_A"], label="Product B")
plt.bar(
    df["Month"],
    df["Product_C"],
    bottom=df["Product_A"] + df["Product_B"],
    label="Product C"
)

plt.title("Product Wise Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

print("\nDashboard Generated Successfully!")
```
